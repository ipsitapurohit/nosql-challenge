## Notebook Set Up

```python
from pymongo import MongoClient
import pandas as pd
from pprint import pprint
from bson.decimal128 import Decimal128
```

```python
# Create an instance of MongoClient
mongo = MongoClient(port=27017)
```

```python
# assign the uk_food database to a variable name
db = mongo['uk_food']
```

```python
# review the collections in our database

# Get the collection names in the database
collection_names = db.list_collection_names()

# Print the collection names
print("Collections in the uk_food database:")
for collection_name in collection_names:
    print(collection_name)
```

```python
# assign the collection to a variable
establishments = db['establishments']
```

## Part 3: Exploratory Analysis
Unless otherwise stated, for each question: 
* Use `count_documents` to display the number of documents contained in the result.

### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20
hygiene_20 = db.establishments.find({"HygieneScore": 20})

# Use count_documents to display the number of documents in the result
hygiene_20_count = db.establishments.count_documents({"HygieneScore": 20})
print("Number of establishments with a hygiene score of 20:", hygiene_20_count)

# Display the first document if it exists
hygiene_20_documents = list(hygiene_20)

if hygiene_20_documents:
    pprint(hygiene_20_documents[0])
else:
    print("No documents found with a hygiene score of 20")
```

```python
# Convert the result to a Pandas DataFrame
df_hygiene_20 = pd.DataFrame(hygiene_20_documents)

# Display the number of rows in the DataFrame
print("Number of rows in DataFrame:", len(df_hygiene_20))

# Display the first 10 rows of the DataFrame
print("First 10 rows:")
print(df_hygiene_20.head(10))
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
london_regex = ".*London.*"  # Using regex to match variations of London's Local Authority name
london_establishments_cursor = db.establishments.find({"LocalAuthorityName": {"$regex": london_regex, "$options": "i"}, "RatingValue": {"$gte": 4}})
london_establishments_count = db.establishments.count_documents({"LocalAuthorityName": {"$regex": london_regex, "$options": "i"}, "RatingValue": {"$gte": 4}})
print("Number of establishments in London with RatingValue >= 4:", london_establishments_count)

# Use count_documents to display the number of documents in the result

# Display the first document in the results using pprint
london_establishments = list(london_establishments_cursor)
if london_establishments:
    print("\nFirst document in the results:")
    pprint(london_establishments[0])
else:
    print("No establishments found in London with RatingValue >= 4")
```

```python
# Convert the result to a Pandas DataFrame
df_london_establishments = pd.DataFrame(london_establishments)

# Display the number of rows in the DataFrame
print("Number of rows in DataFrame:", len(df_london_establishments))

# Display the first 10 rows of the DataFrame
print("First 10 rows:")
print(df_london_establishments.head(10))
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Rating value must equal 5
# Sort by hygiene score

from decimal import Decimal

# Find the document for "Penang Flavours"
penang_flavours_document = db.establishments.find_one({"BusinessName": "Penang Flavours"})

# Search within 0.01 degree on either side of the latitude and longitude.
degree_search = 0.01
latitude_decimal = penang_flavours_document["geocode"]["latitude"].to_decimal()
longitude_decimal = penang_flavours_document["geocode"]["longitude"].to_decimal()

# Convert Decimal to float
latitude = float(latitude_decimal)
longitude = float(longitude_decimal)

# Define the query to find establishments with RatingValue of 5 within the search area
query = {
    "RatingValue": 5,
    "Latitude": {"$gte": latitude - degree_search, "$lte": latitude + degree_search},
    "Longitude": {"$gte": longitude - degree_search, "$lte": longitude + degree_search}
}

sort = [("scores.Hygiene", 1)]  # 1 for ascending order

limit = 5

top_5_establishments = db.establishments.find(query).sort(sort).limit(limit)

# Print the results
print("Top 5 establishments with RatingValue of 5, sorted by lowest hygiene score, nearest to Penang Flavours:")
for establishment in top_5_establishments:
    pprint(establishment)
```

```python
# Convert result to Pandas DataFrame
df_top_5_establishments = pd.DataFrame(top_5_establishments)

# Print the DataFrame
print(df_top_5_establishments)
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create the pipeline
pipeline = [
    # Match establishments with a hygiene score of 0
    {"$match": {"scores.Hygiene": 0}},
    # Group the matches by Local Authority
    {"$group": {"_id": "$LocalAuthorityName", "count": {"$sum": 1}}},
    # Sort the matches from highest to lowest
    {"$sort": {"count": -1}},
    # Limit to the top ten results
    {"$limit": 10}
]

try:
    # Execute the pipeline
    result = db.establishments.aggregate(pipeline)

    # Convert the cursor to a list
    result_list = list(result)

    # Print the number of documents in the result
    result_count = len(result_list)
    print("Number of documents in the result:", result_count)

    if result_count > 0:
        # Print the first 10 results
        print("Top 10 Local Authority areas with establishments having a hygiene score of 0:")
        for doc in result_list:
            print(doc)

        # Convert the result to a Pandas DataFrame
        df_result = pd.DataFrame(result_list)

        # Display the number of rows in the DataFrame
        print("Number of rows in DataFrame:", len(df_result))

        # Display the first 10 rows of the DataFrame
        print("First 10 rows of DataFrame:")
        print(df_result.head(10))
    else:
        print("No establishments found with a hygiene score of 0.")

except Exception as e:
    print("An error occurred:", e)
```

```python
# Convert the result to a Pandas DataFrame
df_result = pd.DataFrame(result_list)

# Display the number of rows in the DataFrame
print("Number of rows in DataFrame:", len(df_result))

# Display the first 10 rows of the DataFrame
print("First 10 rows of DataFrame:", df_result.head(10))
```
