2. Find the BusinessTypeID for "Restaurant/Cafe/Canteen" and return only the `BusinessTypeID` and `BusinessType` fields.

```python
# Find the BusinessTypeID for "Restaurant/Cafe/Canteen" and return only the BusinessTypeID and BusinessType fields

# Define the query to find the BusinessTypeID for "Restaurant/Cafe/Canteen"
query = {"BusinessType": "Restaurant/Cafe/Canteen"}

# Define the projection to return only the BusinessTypeID and BusinessType fields
projection = {"_id": 0, "BusinessTypeID": 1, "BusinessType": 1}

# Find the document matching the query and return only specified fields
result = establishments.find_one(query, projection)

# Print the result
print("BusinessTypeID for 'Restaurant/Cafe/Canteen':")
print(result)
```

3. Update the new restaurant with the `BusinessTypeID` you found.

```python
# Update the new restaurant with the correct BusinessTypeID

# Define the query to find the BusinessTypeID for "Restaurant/Cafe/Canteen"
query = {"BusinessType": "Restaurant/Cafe/Canteen"}

# Define the projection to return only the BusinessTypeID field
projection = {"_id": 0, "BusinessTypeID": 1}

# Find the document matching the query and return only specified fields
result = establishments.find_one(query, projection)

# Extract the BusinessTypeID from the result
business_type_id = result["BusinessTypeID"]

# Define the filter to identify the new restaurant document
filter_query = {"BusinessName": "Penang Flavours"}

# Define the update operation to set the BusinessTypeID
update_operation = {"$set": {"BusinessTypeID": business_type_id}}

# Update the new restaurant document with the BusinessTypeID
update_result = establishments.update_one(filter_query, update_operation)

# Check if the update was successful
if update_result.modified_count > 0:
    print("New restaurant updated successfully with BusinessTypeID.")
else:
    print("No document was updated.")
```

```python
# Confirm that the new restaurant was updated

# Define the query to find the updated restaurant
query = {"BusinessName": "Penang Flavours"}

# Find and retrieve the updated restaurant document
updated_restaurant = establishments.find_one(query)

# Print the updated restaurant document
if updated_restaurant:
    print("Updated Restaurant:")
    print(updated_restaurant)
else:
    print("Restaurant not found.")
```

5. Some of the number values are stored as strings, when they should be stored as numbers.

Use `update_many` to convert `latitude` and `longitude` to decimal numbers.

```python
# Change the data type from String to Decimal for longitude and latitude

# Import Decimal class from the decimal module
from bson.decimal128 import Decimal128
```

```python
# Define the update operation to change data type for longitude and latitude
update_operation = [
    {"$set": {"geocode.longitude": {"$toDecimal": "$geocode.longitude"}}},
    {"$set": {"geocode.latitude": {"$toDecimal": "$geocode.latitude"}}}
]

# Update all documents to change data type for longitude and latitude
update_result = establishments.update_many({}, update_operation)

# Print the number of documents updated
print("Number of documents updated:", update_result.modified_count)
```

Use `update_many` to convert `RatingValue` to integer numbers.

```python
# Set non 1-5 Rating Values to Null
non_ratings = ["AwaitingInspection", "Awaiting Inspection", "AwaitingPublication", "Pass", "Exempt"]
establishments.update_many({"RatingValue": {"$in": non_ratings}}, [ {'$set':{ "RatingValue" : None}} ])
```

```python
# Change the data type from String to Integer for RatingValue

# Define the update operation to convert RatingValue to integer numbers
update_operation = [
    {"$set": {"RatingValue": {"$toInt": "$RatingValue"}}}
]

# Update all documents to change data type for RatingValue
update_result = establishments.update_many({}, update_operation)

# Print the number of documents updated
print("Number of documents updated:", update_result.modified_count)
```

```python
# Check that the coordinates and rating value are now numbers

# Retrieve a sample document from the collection
sample_document = establishments.find_one()

print(sample_document)
```
